```python
col = ['Street.Address', 'City', 'State', 'Postcode','Longitude', 'Latitude', 'CT','CB']
df_new = pd.DataFrame([], columns = col)
point = 0 
for i in range(len(df)):
    ads = df.iloc[i,5]
    cy = df.iloc[i,6]
    st = df.iloc[i,7]
    zp  = df.iloc[i,9]
    lon = df.iloc[i,12]
    lat = df.iloc[i,13]
    s = cg.address(ads, city=cy, state=st, zipcode=zp)
    if s == []:
        cb_id = 'NA'
        ct_id = 'NA'
        df_new = df_new.append({'Street.Address': ads,'City': cy, 'State': st, 'Postcode': zp, 'Longitude': lon,'Latitude': lat, 'CT': ct_id,'CB' : cb_id}, ignore_index=True)
    else:
        cb_id = s[0]['geographies']['2010 Census Blocks'][0]['GEOID']
        ct_id = cb_id[:11]
        df_new = df_new.append({'Street.Address': ads,'City': cy, 'State': st, 'Postcode': zp, 'Longitude': lon,'Latitude': lat, 'CT': ct_id,'CB' : cb_id}, ignore_index=True)
    point += 1
    print('Number:', point)
```
