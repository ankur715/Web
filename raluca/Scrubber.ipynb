# Scrubber
## _cleaning by adding Freight Charge and Total Revenue of R1 and AC HWBs to original_

```python
import pandas as pd

s = pd.read_excel("Sample10 112819.xls")
s
```

```python
# new object without R1 and AC rows
# new object of the R1 and AC rows
noR1AC = s[~s['HWB'].str.contains('R1') & ~s['HWB'].str.contains('AC')]
R1AC = s[~s['HWB'].str.contains('R1') | ~s['HWB'].str.contains('AC')]
# noR1AC
# R1AC
```

```python
# getting boolean of indices for R1 AC
s['HWB'].str.contains('R1') | s['HWB'].str.contains('AC')
```

```python
# getting indices
R1ACindex = s[s['HWB'].str.contains('R1') | s['HWB'].str.contains('AC')].index
R1ACindex
```

```python
# replace with empty, then check if == to first
s['HWB'][1].replace('R1', '')
# s['HWB'][1].replace('R1', '') == s['HWB'][0]
```

```python
# select string except for last 2 characters
s['HWB'][1][:-2]
```

```python
# another option to remove R1 and AC
s['HWB'].str.strip("R1")
s['HWB'].str.strip("AC")
```

```python
R1ACindex
```

```python
# adding the Freight Charge and Total Revenue of R1 AC to original 
# The comments are the steps I took to detect/diagnose errors 
for i in range(len(s)):
    for j in range(len(R1ACindex)):
#         print(s["HWB"][i])
#         print(R1ACindex[j])
#         print(s[R1ACindex[j]])
#         print("AAAAAAAAAAA", s["HWB"][R1ACindex[j]])
        if s["HWB"][i] == s["HWB"][R1ACindex[j]]:
#             print(s["HWB"][i])
#             print(s["Freight Charge"][i])
            s["Freight Charge"][i-1] = s["Freight Charge"][i-1] + s["Freight Charge"][i]
            s["Total Revenue"][i-1] = s["Total Revenue"][i-1] + s["Total Revenue"][i]
#     print(s["Freight Charge"][i])
```

```python
# removing the R1 and AC since they're added to original already
s.drop(R1ACindex)
s = s.drop(R1ACindex)
s
```

```python
count = s[["HWB","Freight Charge","Total Revenue"]].groupby("HWB").agg(['count','mean','sum']).iloc[:,0]
FT = s[["HWB","Freight Charge","Total Revenue"]].groupby("HWB").agg(['mean','sum']).iloc[:,:]
pd.merge(count, FT, on="HWB")
```
